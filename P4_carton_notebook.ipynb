Première classe dont le but est de calculer et d'afficher un graphique représentant la proportion de personne en sous-nutrition dans le monde et par pays pour l'année 2017

```python
class Personne_en_sn:
    def __init__(self, sous_nutrition, population):
        #selection des colones pays et popenmillion pour l'année 2016-2018
        self.sn = sous_nutrition[sous_nutrition['Année'] == '2016-2018'].filter(items=['Pays','PopEnMillion'])
        #selection de population pour l'année 2017
        self.pop = population[population['Année'] == 2017]
        
    def _calcul(self):
        #fusion des tables sur la colone pays
        self.requete = pd.merge(self.pop, self.sn, on="Pays")
        #on supprime les nan
        self.requete.dropna(inplace = True)
        #calcul de la proportion de personne en sous_alimentation
        self.requete['Proportion'] = self.requete['PopEnMillion']  / (self.requete['PopEnMillier'] / 1000.0)*100.0
        self.requete['Proportion'] = np.round(self.requete['Proportion'], decimals=2)
        #calcul de la propotion de personne en sous nutrition au niveau mondial
        self.snmondial = self.sn['PopEnMillion'].sum()  / (self.pop['PopEnMillier'].sum() / 1000.0)*100.0
        self.snmondial = round(self.snmondial,2)
        
    def affichage_resultat(self):
        self._calcul()        
        self.requete = self.requete.sort_values(by="Proportion")
        
        fig, ax = plt.subplots(figsize=(10, 20))
        ax.barh(self.requete['Pays'], self.requete['Proportion'], align='center',color=sns.color_palette("coolwarm",len(self.requete['Proportion'])))
        for i ,v in enumerate(self.requete['Proportion']):
            ax.text(v , i, str(v)+'%')
        ax.axvline(x=self.snmondial, linewidth=2, color='r')
        text = str(self.snmondial) + '% de la population mondial'
        ax.text(self.snmondial +1,10,text, va='center', fontsize=28, color='r')
        ax.text(self.snmondial +3,8,'en sous nutrition', va='center', fontsize=28, color='r')
        ax.set_xlabel('Sous nutrition en %')
        ax.set_title('Proportion de la sous-nutrition par pays')
        print(self.snmondial)
        
p=Personne_en_sn(sous_nutrition, population)
p.affichage_resultat()
```

Dans la classe suivante, on va étudier l'utilisation de la disponibilité intérieure mondiale, la part consommé en nourriture humaine, la part en nourriture animal et la part perdu, on affichera un diagramme circulaire pour afficher cette proportion.

```python
class Etude_disponibilite:
    def __init__(self, dispo_alimentaire):
        
        dispo = dispo_alimentaire.filter(items=['Pays','Aliments pour animaux',
                                                     'Disponibilité intérieure','Nourriture','Pertes'])
        self.table = dispo.fillna(0)

    def dispo_intérieur(self):
        #cette colone est créer pour avoir un total de 100% c'est une synthése des champs semences,
        #perte, autres utilisations, traitement
        self.table['Autres'] = self.table['Disponibilité intérieure'] - self.table['Nourriture']  - self.table['Pertes'] - self.table['Aliments pour animaux']
                
        #calcul des proportions pour le diagramme circulaire
        Autres = self.table['Autres'].sum() / self.table['Disponibilité intérieure'].sum()
        Nourriture = self.table['Nourriture'].sum() / self.table['Disponibilité intérieure'].sum()
        Pertes = self.table['Pertes'].sum() / self.table['Disponibilité intérieure'].sum()
        Animaux = self.table['Aliments pour animaux'].sum() / self.table['Disponibilité intérieure'].sum()
        
        labels = 'Autres', 'Nourriture', 'Pertes', 'Aliments pour animaux'
        liste = [Autres, Nourriture, Pertes, Animaux]
        fig1, ax1 = plt.subplots(figsize=(5, 5))
        ax1.pie(liste, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
        ax1.axis('equal')  
        ax1.set_title('Disponibilité alimentaire mondial par usage', fontsize=20)


e = Etude_disponibilite(dispo_alimentaire)
e.dispo_intérieur()
```

Dans la classe Etude_pays on va afficher deux graphiques, un sur l'aide alimentaire par pays et l'autre plus affinés en prenant l'aide alimentaire par pays et par habitants.

```python
class Etude_pays:
    def __init__(self, aide_alimentaire, population, dispo_alimentaire, sous_nutrition):
        self.aide = aide_alimentaire
        self.pop = population[population['Année'] == 2017]

        
    def aide_alimentaire(self):
        #on somme le total de l'aide alimentaire pour chaque pays
        self.requete = self.aide.groupby('Pays')[['QteEnTonne']].sum()
        self.requete = self.requete.sort_values(by='QteEnTonne' , ascending=True)
        self.requete = self.requete.reset_index()
        #on affiche le graphique
        fig, ax = plt.subplots(figsize=(10, 20))
        ax.barh(self.requete['Pays'], self.requete['QteEnTonne'],color=sns.color_palette("coolwarm",len(self.requete['QteEnTonne'])))
        for i ,v in enumerate(self.requete['QteEnTonne']):
            ax.text(v , i, str(v) + ' tonnes')
        ax.set_xlabel('aide alimentaire en tonnes')
        ax.set_title('Aides alimentaires par pays, cummule depuis 2013')
        
    def aide_alimentaire_par_habitant(self):
        self.requete = pd.merge(self.requete, self.pop, on='Pays')
        #je laisse la population en millier ce qui donnera la qté par habitant en kg au lieu de tonne
        self.requete['QteParHabitant'] = round(self.requete['QteEnTonne'] / self.requete['PopEnMillier'],2)
        self.requete = self.requete.sort_values(by='QteParHabitant' , ascending=True)
        self.requete = self.requete.reset_index()
        # affichage du graphique
        fig, ax = plt.subplots(figsize=(10, 20))
        ax.barh(self.requete['Pays'], self.requete['QteParHabitant'],color=sns.color_palette("coolwarm",len(self.requete['QteParHabitant'])))
        for i ,v in enumerate(self.requete['QteParHabitant']):
            ax.text(v , i, str(v) + ' kg')
        ax.set_xlabel('aide alimentaire en kilogrammes par habitant')
        ax.set_title('Aides alimentaires par pays, cummule depuis 2013 en kg / habitant')

        
ep =Etude_pays( aide_alimentaire, population, dispo_alimentaire, sous_nutrition)
ep.aide_alimentaire()
```

Dans cette classe, je vais afficher la disponibilité alimentaire par pays et par habitants, pour voir les 10 pays les plus fragiles et les 10 plus solides. Je ferais aussi un petit tableau pour voir en détail les 20 pays les plus fragiles.

```python
class Dispo_par_pays_habitant:
    def __init__(self,dispo_alimentaire):
        self.dispo_import = dispo_alimentaire
        self.dispo = dispo_alimentaire.filter(items=['Pays', 'Dispo alimentaire kcal/p/J'])
        
    def disponibilite(self):
        #on fait la somme de la disponibilité alimentaire par pays en kcal
        self.dispo = self.dispo.groupby('Pays')[['Dispo alimentaire kcal/p/J']].sum()
        self.dispo = self.dispo.sort_values(by='Dispo alimentaire kcal/p/J' , ascending=True)
        
        #on récupère les 10 dernières et 10 premières lignes pour montrer les faibles et les forts
        top = self.dispo[-10:]
        flop = self.dispo[:10]
        top = top.reset_index()
        flop = flop.reset_index()
        #ensuite je créer un double graphique pour une mise en parallele 
        plt.figure(figsize=(10, 10))
        
        ax1 = plt.subplot(2, 1, 1)
        
        plt.title("Les 10 pays où la disponibilité alimentaire est la plus faible", fontsize = 16)
        ax1.set_xlim(1500, 4000)
        ax1.axvline(x=2500, linewidth=2, color='r')
        plt.barh(flop['Pays'], flop['Dispo alimentaire kcal/p/J'])
        text = 'Un homme adulte à besoin de 2500kcal par jour pour vivre'
        ax1.text(2550,5,text, va='center', color='r')
        plt.xlabel('Disponibilité en kcal par jour et par habitant pour manger à sa faim')

        ax2 = plt.subplot(2, 1, 2, sharex = ax1)
        plt.title("Les 10 pays où la disponibilité alimentaire est la plus forte", fontsize = 16)
        ax2.axvline(x=2500, linewidth=2, color='r')
        plt.barh(top['Pays'], top['Dispo alimentaire kcal/p/J'])
        plt.xlabel('Disponibilité en kcal par jour et par habitant')
        plt.tight_layout() 
        
    def importation(self):
        #j'importe et somme ce qui me sera utile
        requete = self.dispo_import.groupby('Pays')[['Dispo alimentaire kcal/p/J','Importations','Production','Exportations',
                                                     'Disponibilité intérieure','Nourriture' ]].sum()
        #je garde les pays qui sont sur la limite de 2500 à 2800kcal
        requete = requete[(requete['Dispo alimentaire kcal/p/J']> 2500) & (requete['Dispo alimentaire kcal/p/J'] < 2800)]
        #je calcul un ratio production / disponibilité intérieur, je ne rentre pas plus dans le détail
        requete['Rapport disponibilite sur prod'] = requete['Production'] / requete['Disponibilité intérieure']
        #à partir du ratio précédent je calcul la part produite en interne pour l'affichage du graphique
        requete['Dispo alimentaire kcal/p/J produit'] = requete['Rapport disponibilite sur prod'] * requete['Dispo alimentaire kcal/p/J']
        #on tri et garde les 10 valeurs les plus basses
        requete = requete.sort_values(by='Rapport disponibilite sur prod' , ascending=True)
        requete = requete[:10]
        requete = requete.reset_index()
        #afficahge du graphique
        sns.set_theme(style="whitegrid")
        f, ax = plt.subplots(figsize=(10, 5))
     
        sns.set_color_codes("pastel")
        sns.barplot(x="Dispo alimentaire kcal/p/J", y="Pays", data=requete, label="Disponibilité alimentaire total",
                    color="b")
        sns.set_color_codes("muted")
        sns.barplot(x="Dispo alimentaire kcal/p/J produit", y="Pays", data=requete, label="Disponibilité alimentaire produit localement", color="b")
        ax.legend( frameon=True)
        ax.set_title("Les pays avec une disponibilité suffisante, mais dépendant des importations", fontsize = 16)
    
    def vue_détaillé(self):
        #on somme par pays les données de disponibilité nutritionelles
        detaille = self.dispo_import.groupby('Pays')[['Dispo alimentaire kcal/p/J',
                                                      'Matière grasse g/p/j',
                                                      'Protéines g/p/j']].sum()
        detaille = detaille.sort_values(by='Dispo alimentaire kcal/p/J' , ascending=True)
        detaille = detaille.reset_index()
        detaille = detaille[:20]
        #je personnalise l'affichage du dataframe pour une mise en valeur des cas difficile
        cm = sns.light_palette("red",reverse=True, as_cmap=True)
        display(detaille.style.background_gradient(cmap=cm).set_precision(2).set_caption("Les 20 pays les plus fragiles"))
        
dph = Dispo_par_pays_habitant(dispo_alimentaire) 
dph.disponibilite()
```

Dans cette classe on va voir ce qui se passerait pour la Thaïlande avec et sans exportation du Manioc.

```python
class Thailande:
    def __init__(self, dispo_alimentaire, population):
        dispo = dispo_alimentaire[dispo_alimentaire['Pays'] == 'Thaïlande']
        pop = population[(population['Pays'] == 'Thaïlande') & (population['Année'] == 2017)]
        self.requete = pd.merge(dispo, pop, on='Pays')
        
    def _calcul(self):
        #dans cette fonction on calcul la disponibilite alimentaire en kcal / j / h, si le manioc n'était plus exporté
        manioc = self.requete[self.requete['Produit'] == 'Manioc'] 
        valeur_nutritionnel_manioc = float(manioc['Exportations'] * manioc['Dispo alimentaire kcal/p/J'] / manioc['Nourriture'])
        return round(valeur_nutritionnel_manioc,2)
    
    def analyse(self):
        avant = self.requete.groupby('Pays')[['Dispo alimentaire kcal/p/J']].sum()
        
        display(avant.style.set_caption("avec export du manioc :"))
        
        manioc = self._calcul()
        apres = self.requete
        #on met à jour la colone disponibilite alimentaire avec la nouvelle valeur que l'on a calculé
        apres.loc[apres['Produit'] =='Manioc', 'Dispo alimentaire kcal/p/J'] = manioc
        apres = apres.groupby('Pays')[['Dispo alimentaire kcal/p/J']].sum()
        display(apres.style.set_caption("sans export du manioc :"))
        
thai = Thailande(dispo_alimentaire, population)
thai.analyse()
```

Comme une grosse partie de la consommation de manioc est pour la nourriture animale et au vu des questions précédentes, il serait intéressant de voir ce qui se passerait si la planète passait sur un régime Végan (sans exploitation animale)

```python
class Vegan:
    def __init__(self, dispo_alimentaire, population):
        self.pop = population[population['Année'] == 2017]
        #par simplification on garde que ce qui est vegetale
        self.vege = dispo_alimentaire[dispo_alimentaire['Origine'] == 'vegetale']
        self.vege = self.vege.fillna(0)

        
    def veg(self):
        #on calcul la disponibilité alimentaire si on change de régime en ajoutant la part normalement prévu
        #pour de l'élevage
        self.vege['Vegan kcal/p/J'] = self.vege['Dispo alimentaire kcal/p/J'] + (self.vege['Aliments pour animaux'] * self.vege['Dispo alimentaire kcal/p/J'] / self.vege['Nourriture']).replace(np.inf, 0)
        self.vege['Vegan protéines'] = self.vege['Protéines g/p/j'] + (self.vege['Aliments pour animaux'] * self.vege['Protéines g/p/j'] / self.vege['Nourriture']).replace(np.inf, 0)
        self.vege['Vegan graisses'] = self.vege['Matière grasse g/p/j'] + (self.vege['Aliments pour animaux'] * self.vege['Matière grasse g/p/j'] / self.vege['Nourriture']).replace(np.inf, 0)

        requete = self.vege.groupby(['Pays'])[['Vegan kcal/p/J','Vegan protéines','Vegan graisses']].sum()
       
        #on joint la table population
        requete = pd.merge(requete, self.pop, on='Pays')
      
        #on calcul le nbr de calorie, graisse, proteine total par pays
        requete['total_calorie'] = requete['PopEnMillier'] * 1000 * requete['Vegan kcal/p/J']
        requete['total_graisse'] = requete['PopEnMillier'] * 1000 * requete['Vegan graisses']
        requete['total_protéine'] = requete['PopEnMillier'] * 1000 * requete['Vegan protéines']
        #on fait la somme des calories dans le monde et on divise par 2500
        nbr_p_nourissablec = int(round(requete['total_calorie'].sum() / 2500,0))
        #on fait la somme des protéines dans le monde et on divise par 60
        nbr_p_nourissablep = int(round(requete['total_protéine'].sum() / 60,0))
        #on fait la somme des graisses dans le monde et on divise par 60
        nbr_p_nourissableg = int(round(requete['total_graisse'].sum() / 60,0))
        
        print("Avec un régime végan sur la base des calories, il est possible de nourrir",f"{nbr_p_nourissablec:,d}", "d'habitants")
        print("Avec un régime végan sur la base des protéines, il est possible de nourrir",f"{nbr_p_nourissablep:,d}", "d'habitants")
        print("Avec un régime végan sur la base des lipides, il est possible de nourrir",f"{nbr_p_nourissableg:,d}", "d'habitants")
        
regime_vegan = Vegan(dispo_alimentaire, population)
regime_vegan.veg()
```

Un régime végan sans modification des cultures est trop pauvre en matière grasse pour nourrir la population. Dans la classe suivante, on va analyser les qualités nutritionnelles des céréales.

```python
class qualite_nutritionnel:
    def __init__(self, population, dispo_alimentaire):
        pop = population[population['Année'] == 2017]
        dispo = dispo_alimentaire
        self.requete = pd.merge(dispo, pop, on='Pays')
        
    def calcul(self):
        # la liste des céréales
        liste_cereal = ['Blé','Céréales, Autres', 'Maïs' ,'Orge', 'Avoine', 'Seigle', 'Millet']
        #création d'un data frame vite pour stocker nos résultats
        df_cereal = pd.DataFrame()
        
        #cette boucle va remplir le data frame en annalysant pour chaque céréale les valeurs nutritives
        for cereal in liste_cereal: 
            r_cereal = self.requete[self.requete["Produit"] == cereal]
            #je prends la valeur la plus élevé, pour éviter l'imprécision des petites valeurs
            r_cereal = r_cereal.loc[r_cereal["Dispo alimentaire kg/p/an"].idxmax()]
            #calcul pour 100 grammes (ce qui me permet de comparer avec ce que j'ai trouver sur le net)
            mg = (r_cereal['Matière grasse g/p/j'] * 365) * 100.0 / (r_cereal['Dispo alimentaire kg/p/an']*1000)
            pr = (r_cereal['Protéines g/p/j'] * 365) * 100.0 / (r_cereal['Dispo alimentaire kg/p/an']*1000)
            
            cal = (r_cereal['Dispo alimentaire kcal/p/J']*365 )*100 / (r_cereal['Dispo alimentaire kg/p/an']*1000)
            df_cereal = df_cereal.append({'Céréale' : cereal, 'Matière grasse' : mg, 'Protéines' : pr,
                                         'kcal' : cal}, ignore_index = True)

        g=sns.catplot(x="Céréale", y="Matière grasse", kind="bar", data=df_cereal).set(title='Apport en matière grasse pour 100gr')
        g.set_xticklabels( rotation=30)
        e=sns.catplot(x="Céréale", y="Protéines", kind="bar", data=df_cereal).set(title='Apport en protéines pour 100gr')
        e.set_xticklabels( rotation=30)
        f=sns.catplot(x="Céréale", y="kcal", kind="bar", data=df_cereal).set(title='Apport en kcal pour 100gr')
        f.set_xticklabels( rotation=30)
        return df_cereal
    
qn = qualite_nutritionnel(population , dispo_alimentaire)
analyse_cereal = qn.calcul()
```

Le maïs est la céréale la plus riche, mais c'est aussi celle qui consomme le plus d'eau, donc un meilleur choix serait l'avoine, car c'est celle apportant le plus de graisse et l'on a vu que c'était le facteur limitant, tout en étant économe en eau. Dans la classe qui suit, je vais remplacer toutes les céréales par de l'avoine, pour voir s'il serait possible de nourrir plus de population.

```python
class Solution_possible:
    def __init__(self, analyse_cereal, dispo_alimentaire):

        #on garde que ce qui est vegetale
        self.vege = dispo_alimentaire[dispo_alimentaire['Origine'] == 'vegetale']
        self.cereal = analyse_cereal

    def _calcul(self):
        # on calcul une valeur qui permettra de retrouver la colone disponibilité "xxx / personne / jour" 
        #en multipliant la colone disponibilité alimentaire en kg
        graisse = float(analyse_cereal['Matière grasse'][analyse_cereal['Céréale'] == 'Avoine']) * 10 / 365 
        proteine = float(analyse_cereal['Protéines'][analyse_cereal['Céréale'] == 'Avoine']) * 10 / 365 
        kcal = float(analyse_cereal['kcal'][analyse_cereal['Céréale'] == 'Avoine']) * 10 / 365 
        
        list_cereal = ['Blé', 'Seigle', 'Orge', 'Millet', 'Céréales, Autres', 'Maïs']
        for cereale in list_cereal:
            #dans cette boucle on met à jour les colones matiere grasse, proteine, dispo alimentaire avec les nouvelles valeurs
            self.vege.loc[self.vege['Produit'] == cereale, 'Matière grasse g/p/j'] = self.vege['Dispo alimentaire kg/p/an'] * graisse
            self.vege.loc[self.vege['Produit'] == cereale, 'Protéines g/p/j'] = self.vege['Dispo alimentaire kg/p/an'] * proteine
            self.vege.loc[self.vege['Produit'] == cereale, 'Dispo alimentaire kcal/p/J'] = self.vege['Dispo alimentaire kg/p/an'] * kcal
  
    def solution(self, population):
        self._calcul()
        #on rappelle la classe que l'on avait utilisé pour calculer le régime vegan sur terre
        #mais cette fois avec l'avoine en rempalcement de toutes les autres céréales
        regime_vegan = Vegan(self.vege, population)
        regime_vegan.veg()
        pop = int(population.loc[population['Année'] == 2017,'PopEnMillier'].sum()*1000)
        print("Population en 2017 : ",f"{pop:,d}")
        
        
conclusion = Solution_possible(analyse_cereal, dispo_alimentaire)
conclusion.solution(population)
```
